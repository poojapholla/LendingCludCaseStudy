```python
#Reading the loan.csv file from the local folder using pandas.

loans_df = pd.read_csv('/Users/z004ngl/Desktop/loan.csv')
```

```python
#checkinh for size of columns 

loans_df.columns.size
```

```python
#Filtering out all the columns which are NAN, since this will not add any value to our data analysis.

loans_df = loans_df.dropna(axis=1, how="all")
```

```python
#Checking size of the columns after filtering.

loans_df.columns.size
```

```python
loans_df.head()
```

```python
#check the total rows and columns with shape 
loans_df.shape
```

```python
#understand all the columns data types
loans_df.info(verbose=True)
```

```python
#Find all the columns which has null row values > finding the sum of rows withh nulls for each of the columns.

loans_df.isnull().sum()
```

```python
#Percentage of missing values for all columns 

loans_df.index
```

```python
null_percentage = round(100*(loans_df.isnull().sum() / len(loans_df.index)) , 2)
```

```python
null_percentage
```

```python
#removing all the columns with null percentage greater than 30, since it doesnt add any value to analysis

loans_df_after_nulls_removal =  null_percentage[null_percentage <= 30]
loans_df_after_nulls_removal
```

```python
loans_df_without_nulls = ([null_percentage[null_percentage >= 30.00].index])
loans_df.drop(labels=loans_df_without_nulls[0],axis=1,inplace=True)
```

```python
#checking for rows and columsn after removing columns with nulls greater that 30%, removed 4 columns
loans_df.shape
```

```python
# For rest replacing all the missing values with some values.
#calculating for empl title.


loans_df["emp_title"].head()
```

```python
loans_df.emp_title.mode()
```

```python
#since most the values here is US Army we will fill up the missing data with US Army.
loans_df['emp_title'] = loans_df['emp_title'].fillna('US Army')
```

```python
loans_df.isna().sum()
```

```python
#filling missing values for title
loans_df.title.mode()
```

```python
loans_df['title'] = loans_df['title'].fillna('Debt Consolidation')
```

```python
loans_df["revol_util"]
```

```python
#since for revol util percentage is small we will replace the missing values with 0
#we can also fill this up with median, but since the null rate is very small, filling this up with the median values.
loans_df['revol_util'] = loans_df['revol_util'].fillna(0.0)
```

```python
loans_df["last_pymnt_d"].mode()
```

```python
loans_df['last_pymnt_d'] = loans_df['last_pymnt_d'].fillna('May-16')
```

```python
loans_df["collections_12_mths_ex_med"]
#since mostly the default value is 0.0 we will fill NAN with 0.0
loans_df['collections_12_mths_ex_med'] = loans_df['collections_12_mths_ex_med'].fillna(0.0)
```

```python
loans_df["chargeoff_within_12_mths"]
#since mostly the default value is 0.0 we will fill NAN with 0.0
loans_df['chargeoff_within_12_mths'] = loans_df['chargeoff_within_12_mths'].fillna(0.0)
```

```python
loans_df["pub_rec_bankruptcies"]
#since mostly the default value is 0.0 we will fill NAN with 0.0
loans_df['pub_rec_bankruptcies'] = loans_df['pub_rec_bankruptcies'].fillna(0.0)
```

```python
loans_df["tax_liens"]
#since mostly the default value is 0.0 we will fill NAN with 0.0
loans_df['tax_liens'] = loans_df['tax_liens'].fillna(0.0)
```

```python
loans_df["emp_length"].mode()
loans_df['emp_length'] = loans_df['emp_length'].fillna("10+ years")
```

```python
# checking for all the columns for null values once again after data cleaning and filling data with missing values.
loans_df.isnull().sum()

#Now none of the columsn has missing values.
```
